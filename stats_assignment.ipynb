```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('divorce_netherlands.csv')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=[8, 5])
sns.lineplot(x='Year', y='rate', data =data)
plt.xlabel('Year', size=12)
plt.ylabel('rate(1000s)', size =12)
plt.title('Divore rate 2005 to 2007', size=15)
```

```python
plt.figure(figsize=[8, 5])
sns.barplot(x='Year', y='rate', data=data, palette=['Green'])
plt.xlabel('Year', size=12)
plt.ylabel('divorce rate(1000s)', size=12)
plt.title('Divorce Rate by Year', size =15)
```

```python
data=pd.read_csv('pop_growth.csv')
data.head()
```

```python
plt.gcf().set_size_inches(15, 6)
explode=(0.1,0,0,0)
data['population'].plot(kind='pie',ylabel='',labels=data.type ,autopct='%.0f%%',textprops={'size': 'large'},shadow=True,explode=explode);
plt.title('Population Proportions', size=15)
plt.show()
```

```python
data=pd.read_csv('gender_pop.csv')
data
```

```python
new=pd.melt(data, id_vars='Year', var_name='Gender', value_name='sex')
new
```

```python
plt.figure(figsize=[8, 5])
sns.lineplot(data=new, y='sex',x='Year',hue='Gender',palette =['Red', 'Blue'])
sns.set_style('darkgrid')
plt.yticks([8e6, 8.2e6,8.4e6,8.6e6,8.8e6], ['8M','8.2M','8.4M','8.6M','8.8M'])
plt.xticks([2005,2010,2015,2018], ['2005','2010','2015','2018'])
plt.xlabel('Year', size=12)
plt.ylabel('Population (millions)', size=12)
plt.title('Population Growth by Gender ', size=15)
```
